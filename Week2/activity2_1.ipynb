## คำอธิบายลำดับการทำงาน

### Block A

1. กำหนดค่า `temp = 25` และ `is_raining = True`
2. ตรวจสอบเงื่อนไข `if` แรก: `temp > 20 and not is_raining`
   - `temp > 25` → **True** (25 > 20)
   - `not is_raining` → `not True` → **False**
   - `True and False` → **False** → ข้ามบล็อกนี้ไป
3. ตรวจสอบเงื่อนไข `elif`: `temp <= 20 or is_raining`
   - `temp <= 20` → **False** (25 ไม่น้อยกว่าหรือเท่ากับ 20)
   - `is_raining` → **True**
   - `False or True` → **True** → เข้าบล็อกนี้!
4. **Output ที่ได้:** `Stay indoors or bring an umbrella.`

---

### Block B

1. กำหนด `phrase = "Python 2026 Engine"`
2. ใช้ Set Comprehension วนซ้ำทุกตัวอักษร `c` ใน `phrase`
3. เงื่อนไขคัดกรอง:
   - `c.isalpha()` → เอาเฉพาะตัวอักษร (ไม่เอาตัวเลข/ช่องว่าง)
   - `c.lower() not in 'aeiou'` → ไม่เอาสระ (a, e, i, o, u)
4. ตัวอักษรที่เหลือ (พยัญชนะ) แปลงเป็นตัวพิมพ์เล็กแล้วใส่ใน Set (Set จะไม่มีค่าซ้ำ)
5. **ตัวพยัญชนะที่ได้จาก `"Python 2026 Engine"`:**
   - P, y, t, h, n → p, y, t, h, n
   - E, n, g, n → e ❌ (สระ), n, g, n
   - ผลลัพธ์ใน Set: `{'p', 'y', 't', 'h', 'n', 'g'}`
6. **Output ที่ได้:** `Unique consonants: {'p', 'y', 't', 'h', 'n', 'g'}` *(ลำดับอาจสลับได้เนื่องจากเป็น Set)*

````python
# Block A: Evaluation Check
temp = 25
is_raining = True

if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.") # Expected output matches here
else:
    print("Unsure about the weather.")
````

````python
# Block B: Vectorized Filter
phrase = "Python 2026 Engine"
consonants = {c.lower() for c in phrase if c.isalpha() and c.lower() not in 'aeiou'}
print(f"Unique consonants: {consonants}")
````

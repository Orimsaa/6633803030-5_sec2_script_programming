## เรื่อง: ผจญภัยในป่าลึก 

คุณตื่นขึ้นมากลางป่าลึก ไม่รู้ว่าตัวเองอยู่ที่ไหน มองเห็นทางแยกอยู่สองทาง...

---

## ลำดับการทำงาน (Logic Flow)

```
INPUT: ถามชื่อผู้เล่น

INPUT: เลือกทางเดิน → "ซ้าย" หรือ "ขวา"

IF เลือก "ซ้าย":
    → พบถ้ำมืด
    INPUT: จะเข้าถ้ำหรือไม่? → "ใช่" หรือ "ไม่"
    IF ตอบ "ใช่":
        → พบสมบัติ! จบด้วยชัยชนะ
    ELSE:
        → เดินวนหลงป่า จบด้วยการแพ้

ELIF เลือก "ขวา":
    → พบแม่น้ำกว้าง
    INPUT: จะว่ายน้ำข้ามหรือไม่? → "ใช่" หรือ "ไม่"
    IF ตอบ "ใช่":
        → ข้ามได้พบหมู่บ้าน จบด้วยชัยชนะ
    ELSE:
        → รอจนค่ำ ถูกสัตว์ป่าไล่ จบด้วยการแพ้

ELSE:
    → ตอบไม่ถูกต้อง ให้เลือกใหม่
```

````python
# Activity 2.2: Choose Your Own Adventure - ผจญภัยในป่าลึก
# ชื่อ: อัษฎาวุธ เรือนแก้ว
# รหัสนักศึกษา: 663380303-5

print("=" * 40)
print("   ผจญภัยในป่าลึก ")
print("=" * 40)

name = input("ใส่ชื่อของคุณ: ")
print(f"\nยินดีต้อนรับ {name}! คุณตื่นขึ้นมากลางป่าลึก...")
print("คุณมองเห็นทางแยก 2 ทาง")

choice = input("คุณจะเดินไปทาง 'ซ้าย' หรือ 'ขวา'? ").strip()

if choice == "ซ้าย":
    print("\nคุณพบถ้ำมืดลึกลับ...")
    enter = input("คุณจะเข้าไปในถ้ำหรือไม่? (ใช่/ไม่): ").strip()
    if enter == "ใช่":
        print(f"\n {name} กล้าเข้าถ้ำ และพบสมบัติล้ำค่า! คุณชนะ!")
    else:
        print(f"\n {name} ไม่กล้าเข้า เดินวนหลงป่าทั้งคืน... คุณแพ้")

elif choice == "ขวา":
    print("\nคุณพบแม่น้ำกว้างขวาง...")
    swim = input("คุณจะว่ายน้ำข้ามแม่น้ำหรือไม่? (ใช่/ไม่): ").strip()
    if swim == "ใช่":
        print(f"\n {name} ว่ายข้ามแม่น้ำได้สำเร็จ และพบหมู่บ้าน! คุณชนะ!")
    else:
        print(f"\n {name} รอจนค่ำ ถูกสัตว์ป่าไล่... คุณแพ้")

else:
    print("\n ไม่เข้าใจคำตอบ กรุณาเลือก 'ซ้าย' หรือ 'ขวา' เท่านั้น")

print("\n" + "=" * 40)
print("        จบการผจญภัย!")
print("=" * 40)
````

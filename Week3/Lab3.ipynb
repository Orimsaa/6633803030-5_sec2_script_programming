## Lab 3.1 Part 1: Single Number Multiplication Table
รับค่าจากผู้ใช้ แล้วใช้ลูป `for` ร่วมกับ `range(1, 13)` เพื่อพิมพ์สูตรคูณของเลขนั้นตั้งแต่คูณ 1 ถึง 12

```python
# --- Lab 3.1 Part 1: Single Number Multiplication Table ---
# รับค่าจากผู้ใช้
n = int(input("Enter a number for its multiplication table (e.g., 7): "))

print(f"--- Multiplication Table for {n} ---")
# ใช้ for loop วนจาก 1 ถึง 12 เพื่อพิมพ์สูตรคูณ
for i in range(1, 13):
    print(f"{n} x {i:2d} = {n * i:3d}")

print()  # spacer
```

## Lab 3.1 Part 2: Full 12x12 Multiplication Grid (Challenge)
ใช้ลูป `for` แบบซ้อนกัน (Nested loops) เพื่อพิมพ์ตารางสูตรคูณขนาด 12x12

**หลักการ:**
- **ลูปนอก (Outer loop)**: ใช้ตัวแปร `i` วนแต่ละแถว (row) จาก 1 ถึง 12
- **ลูปใน (Inner loop)**: ใช้ตัวแปร `j` วนแต่ละคอลัมน์ (column) จาก 1 ถึง 12
- ใช้ `f"{i*j:4d}"` เพื่อจัดให้ตัวเลขมีความกว้าง 4 ตำแหน่ง ให้ตารางสวยงามเป็นระเบียบ
- ใช้ `end=""` เพื่อไม่ให้ขึ้นบรรทัดใหม่หลัง print แต่ละตัวเลข
- ใช้ `print()` เปล่าเพื่อขึ้นบรรทัดใหม่หลังจบแต่ละแถว

```python
# --- Lab 3.1 Part 2: Full 12x12 Multiplication Grid ---
print("12x12 Multiplication Table:")
for i in range(1, 13):       # ลูปนอก: วนแต่ละแถว (row 1-12)
    for j in range(1, 13):   # ลูปใน: วนแต่ละคอลัมน์ (column 1-12)
        print(f"{i*j:4d}", end="")  # พิมพ์ผลคูณ จัดให้กว้าง 4 ตำแหน่ง ไม่ขึ้นบรรทัดใหม่
    print()  # ขึ้นบรรทัดใหม่เมื่อจบแต่ละแถว

print()  # spacer
```

## Lab 3.2 Part 1: Countdown Timer
รับค่าเริ่มต้นจากผู้ใช้ ใช้ลูป `while` นับถอยหลังลดลงทีละ 1 จนถึง 0 แล้วพิมพ์ "Blast off!"

**หลักการ:**
- ใช้ `while start >= 0` เป็นเงื่อนไขให้ลูปทำงานจนกว่า start จะน้อยกว่า 0
- ภายในลูป: พิมพ์ค่า `start` ออกมา แล้วลดค่าลง 1 ด้วย `start -= 1`
- หลังออกจากลูป: พิมพ์ "Blast off!"

```python
# --- Lab 3.2 Part 1: Countdown Timer ---
# รับค่าเริ่มต้นจากผู้ใช้
start = int(input("Enter starting number for countdown: "))

# ใช้ while loop นับถอยหลังจนกว่า start จะน้อยกว่า 0
while start >= 0:
    print(start)    # พิมพ์ตัวเลขปัจจุบัน
    start -= 1      # ลดค่าลง 1

print("Blast off!")  # พิมพ์เมื่อนับถอยหลังเสร็จ
```

## Lab 3.2 Part 2: Simple Guessing Game
กำหนดตัวเลขลับ (`secret_number = 42`) ให้ผู้ใช้ทายตัวเลข

**หลักการ:**
- ใช้ `while True` เพื่อให้วนถามซ้ำไปเรื่อยๆ จนกว่าจะทายถูก
- ใช้ `if/elif/else` เพื่อเช็คว่าตัวเลขที่ทาย น้อยไป มากไป หรือถูกต้อง
- ใช้ `break` เพื่อออกจากลูปเมื่อทายถูก

```python
# --- Lab 3.2 Part 2: Simple Guessing Game ---
secret_number = 42  # กำหนดตัวเลขลับ

while True:
    guess = int(input("Guess the number: "))

    if guess < secret_number:        # ถ้าทายน้อยเกินไป
        print(f"{guess} is Too low! Try again.")
    elif guess > secret_number:      # ถ้าทายมากเกินไป
        print(f"{guess} is Too high! Try again.")
    else:                            # ถ้าทายถูก
        print("Congratulations! You guessed it!")
        break  # ออกจากลูป
```

# ⭐ Challenge: Guessing Game with Limited Attempts (while-else)
เพิ่มการจำกัดจำนวนครั้งในการทาย (สูงสุด 5 ครั้ง) โดยใช้ `while` ร่วมกับ `else`

**หลักการของ while-else ใน Python:**
- บล็อก `else` ของ `while` จะทำงาน **เมื่อลูปจบตามปกติ** (เงื่อนไข while เป็น False)
- ถ้าลูปถูกหยุดด้วย `break` บล็อก `else` จะ **ไม่ทำงาน**
- ดังนั้นเราสามารถใช้ `else` เพื่อตรวจสอบว่าผู้เล่นทายไม่ถูกเลยจนโควตาหมด

**เพิ่มเติมจากโจทย์ปกติ:**
- ใช้ `import random` เพื่อสุ่มตัวเลขลับแทนการกำหนดค่าตายตัว
- แสดงจำนวนครั้งที่เหลือให้ผู้เล่นทราบในแต่ละรอบ
- ถ้าทายหมดโควตาจะเฉลยคำตอบให้

```python
# --- Challenge: Guessing Game with Limited Attempts (while-else) ---
import random

# สุ่มตัวเลขลับระหว่าง 1-100 (แทนการกำหนดค่าตายตัว)
secret_number = random.randint(1, 100)
max_attempts = 5      # จำกัดจำนวนครั้งในการทาย
attempts = 0          # ตัวนับจำนวนครั้งที่ทาย

print("═══ Guessing Game (Challenge Mode) ═══")
print(f"I am thinking of a number between 1 and 100.")
print(f"You have {max_attempts} attempts. Good luck!")
print()

# ใช้ while ร่วมกับเงื่อนไข attempts < max_attempts
# ถ้าเกินจำนวนครั้ง -> เงื่อนไขเป็น False -> ลูปจบปกติ -> else ทำงาน
# ถ้าทายถูก -> break -> else ไม่ทำงาน
while attempts < max_attempts:
    attempts += 1
    guess = int(input(f"[Attempt {attempts}/{max_attempts}] Guess: "))

    if guess < secret_number:
        remaining = max_attempts - attempts
        print(f"[Attempt {attempts}/{max_attempts}] {guess} is Too low! Try again. ({remaining} attempts left)")
    elif guess > secret_number:
        remaining = max_attempts - attempts
        print(f"[Attempt {attempts}/{max_attempts}] {guess} is Too high! Try again. ({remaining} attempts left)")
    else:
        print(f"[Attempt {attempts}/{max_attempts}] {guess} → Congratulations! You guessed it in {attempts} attempt(s)!")
        break  # ทายถูก -> ออกจากลูป (else จะไม่ทำงาน)

else:
    # บล็อกนี้ทำงานเฉพาะเมื่อลูป while จบตามปกติ (ไม่ได้ถูก break)
    # หมายความว่าผู้เล่นทายไม่ถูกเลยจนโควตาหมด
    print(f"\nGame Over! You have used all {max_attempts} attempts.")
    print(f"The secret number was: {secret_number}")
```

# ⭐ Challenge: Enhanced Countdown with Progress Bar
ปรับปรุง Countdown Timer ให้สมจริงยิ่งขึ้นโดยใช้ `time.sleep()` เพื่อหน่วงเวลา
และเพิ่ม Progress Bar แสดงสถานะระหว่างนับถอยหลัง

**สิ่งที่เพิ่มเติมจากโจทย์ปกติ:**
- ใช้ `time.sleep(1)` เพื่อหน่วงเวลา 1 วินาทีในแต่ละรอบ ทำให้ดูเหมือนนาฬิกาจับเวลาจริง
- เพิ่ม Input Validation ตรวจสอบว่าผู้ใช้ป้อนตัวเลขมากกว่า 0
- แสดง Progress Bar อย่างง่ายระหว่างนับถอยหลัง

```python
# --- Challenge: Enhanced Countdown with Progress Bar ---
import time

start = int(input("Enter starting number for countdown: "))

# ตรวจสอบว่าผู้ใช้ป้อนตัวเลขมากกว่า 0 (Input Validation)
if start <= 0:
    print("Please enter a number greater than 0!")
else:
    total = start  # เก็บค่าเริ่มต้นไว้คำนวณ progress
    print("=== Countdown Timer (Enhanced) ===")

    while start >= 0:
        # คำนวณ progress bar
        progress = start / total                       # สัดส่วนที่เหลือ (1.0 -> 0.0)
        bar_filled = int(progress * 10)                # จำนวนบล็อกที่เต็ม
        bar_empty = 10 - bar_filled                    # จำนวนบล็อกที่ว่าง
        bar = "█" * bar_filled + "·" * bar_empty       # สร้าง progress bar
        percent = int(progress * 100)                  # เปอร์เซ็นต์

        print(f"Countdown: {start}  [{bar}] {percent:3d}%")
        start -= 1
        time.sleep(1)  # หน่วงเวลา 1 วินาทีให้สมจริง

    print("\n🚀 Blast off!")
```

# ⭐ Challenge: Formatted Multiplication Table with Header
ปรับปรุงตารางสูตรคูณ 12x12 ให้มี **หัวตาราง (Header)** และ **เส้นแบ่ง** เพื่อให้อ่านง่ายยิ่งขึ้น

**สิ่งที่เพิ่มเติมจากโจทย์ปกติ:**
- เพิ่มหัวคอลัมน์ (Column Header) แสดงเลข 1-12 ด้านบน
- เพิ่มหัวแถว (Row Header) แสดงเลขด้านซ้าย
- เพิ่มเส้นแบ่ง (Separator) ระหว่าง header กับข้อมูล
- ใช้ String Formatting ขั้นสูง เพื่อให้ตารางจัดเรียงสวยงาม

```python
# --- Challenge: Formatted Multiplication Table with Header ---
print("═══ Formatted 12x12 Multiplication Table ═══")

# พิมพ์หัวคอลัมน์ (Column Header)
header = "   x |"  # มุมซ้ายบน
for j in range(1, 13):
    header += f"{j:5d}"  # หัวคอลัมน์ 1-12
print(header)

# พิมพ์เส้นแบ่ง
print("-----+" + "-" * 60)

# พิมพ์ข้อมูลแต่ละแถวพร้อมหัวแถว (Row Header)
for i in range(1, 13):
    row = f"{i:4d} |"  # หัวแถว
    for j in range(1, 13):
        row += f"{i*j:5d}"  # ผลคูณ
    print(row)
```

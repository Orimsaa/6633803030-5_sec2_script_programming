# Lab 3.1: โปรแกรมสร้างสูตรคูณ
## Part 1: ตารางสูตรคูณแม่เดียว

```python
n = int(input('Enter a number for its multiplication table (e.g., 7): '))
for i in range(1, 13):
    print(f'{n} x {i} = {n * i}')
print()
```

## Part 2: ตารางสูตรคูณแบบเต็ม (12x12)

```python
print('12x12 Multiplication Table:')
for i in range(1, 13):
    for j in range(1, 13):
        print(f'{i*j:4d}', end='')
    print()
print()
```

# Lab 3.2: โปรแกรมการนับถอยหลัง & เกมทายตัวเลข
## Part 1: โปรแกรมตัวนับถอยหลัง

```python
start = int(input('Enter starting number for countdown: '))
while start >= 0:
    print(start)
    start -= 1
print('Blast off!')
print()
```

## Part 2: เกมทายตัวเลขอย่างง่าย

```python
secret_number = 42
while True:
    guess = int(input('Guess the number: '))
    if guess < secret_number:
        print(f'{guess} is too low! Try again.')
    elif guess > secret_number:
        print(f'{guess} is too high! Try again.')
    else:
        print('Congratulations! You guessed it!')
        break
```

# Activity 3.1
แก้ข้อผิดพลาดของโค้ดที่กำหนดให้

```python
# สมมติว่าโค้ดเดิมมี Error เช่น for i in range(5) print(i)
# โค้ดที่แก้ไขแล้ว:
for i in range(5):
    print(i)
```

# Activity 3.2
ใช้ loop ในการ print() รูปแบบตามที่กำหนด (ตัวอย่างเช่น รูปสามเหลี่ยม)

```python
# ตัวอย่างการใช้ loop พิมพ์รูปสามเหลี่ยม
rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print('*', end='')
    print()
```

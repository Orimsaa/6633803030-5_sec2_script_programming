# lab4_1.py - Student Management System (List Operations)

```python
# lab4_1.py - Student Management System (List Operations)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (ระบบเมนูแบบ while loop) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

print("--- Lab 4.1: Student Management System ---")

# 1) เริ่มต้นด้วย list ว่างสำหรับเก็บชื่อนักศึกษา
student_names = []

# 2) เพิ่มนักศึกษา: รับชื่อจากผู้ใช้ 3 คน แล้ว append() เข้า list
print("\n[1] Add Students")
for i in range(3): #กำหนดตัวแปรใน loop 👈ต้องทำ
    name = input(f"Enter student name #{i + 1}: ") #กำหนดตัวแปร 👈ต้องทำ
    student_names.append(name)

print(f"Current list: {student_names}")

# 3) ค้นหานักศึกษา: ใช้ 'in' เพื่อเช็คว่ามีชื่อนี้อยู่ใน list หรือไม่
#    ถ้าพบ ให้แสดง index ด้วย index()
print("\n[2] Find a Student")
search_name = input("Enter a name to search for: ")

if search_name in student_names: #เขียนเงื่อนไข ถ้าหากมีชื่อในลิส: 👈ต้องทำ
    position = student_names.index(search_name)
    print(f"Found '{search_name}' at index {position}")
else:
    print(f"'{search_name}' not found in the list")

# 4) ลบนักศึกษา: เช็คก่อนว่ามีอยู่จริงหรือไม่ (กัน error) แล้วค่อย remove()
print("\n[3] Remove a Student")
remove_name = input("Enter a name to remove: ")

if remove_name in student_names:
    student_names.remove(remove_name) #เติมฟังก์ชัน remove(remove_name) 👈ต้องทำ
    print(f"'{remove_name}' has been removed")
else:
    print(f"'{remove_name}' not found, nothing removed")

print(f"Updated list: {student_names}")

# 5) เรียงลำดับรายชื่อตามตัวอักษรด้วย sort()
print("\n[4] Sort Students")
student_names.sort() #เติมฟังก์ชันsort()
print(f"Sorted list: {student_names}")

# 6) นับจำนวนนักศึกษาทั้งหมดด้วย len()
print("\n[5] Count Students")
print(f"Total students: {len(student_names)}") #ใช้คำสั่งlen() หาจำนวนนักศึกษาในลิส student_names👈ต้องทำ

# --------------------------------------------------------------------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองเขียนระบบเมนูด้วย while loop (จาก Week 3) ที่ให้ผู้ใช้เลือกได้ว่า
# จะเพิ่ม / ลบ / ค้นหา / เรียงลำดับ / หรือออกจากโปรแกรม
# โดยนำโค้ดแต่ละส่วนด้านบนไปจัดเป็นฟังก์ชันหรือเงื่อนไขในเมนู
# --------------------------------------------------------------------
```

# Challenge Lab 4.1: Student Management System (Menu with While Loop)
โจทย์ท้าทาย: สร้างระบบเมนูสั่งการแบบง่ายๆ โดยใช้ while loop
ให้ผู้ใช้เลือกได้ว่าจะ เพิ่ม / ลบ / ค้นหา / เรียงลำดับ / ออกจากโปรแกรม

```python
# =============================================================
# Challenge Lab 4.1: Student Management System (Menu Version)
# =============================================================
# แนวคิด: นำโค้ดทั้งหมดจาก Lab 4.1 มาจัดใหม่ให้อยู่ในรูปแบบ
# ระบบเมนู (Menu System) โดยใช้ while loop วนซ้ำไปเรื่อยๆ
# จนกว่าผู้ใช้จะเลือก "ออกจากโปรแกรม"
#
# คำสั่ง List ที่ใช้ในโปรแกรมนี้:
#   - append()  : เพิ่มข้อมูลต่อท้าย list
#   - remove()  : ลบข้อมูลตามค่าที่ระบุออกจาก list
#   - index()   : หา index (ตำแหน่ง) ของข้อมูลใน list
#   - sort()    : เรียงลำดับข้อมูลใน list (A-Z)
#   - len()     : นับจำนวนสมาชิกทั้งหมดใน list
#   - in        : ตรวจสอบว่าข้อมูลอยู่ใน list หรือไม่
# =============================================================

print("=" * 50)
print("  Student Management System (Menu Version)")
print("=" * 50)

# สร้าง list ว่างสำหรับเก็บรายชื่อนักศึกษา
student_names = []

# ใช้ while True เพื่อวนลูปแสดงเมนูไปเรื่อยๆ
# โปรแกรมจะหยุดก็ต่อเมื่อผู้ใช้เลือก "5" (ออกจากโปรแกรม)
while True:
    # แสดงเมนูให้ผู้ใช้เลือก
    print("\n--- เมนูหลัก ---")
    print("1. เพิ่มรายชื่อนักศึกษา (Add)")
    print("2. ลบรายชื่อนักศึกษา (Remove)")
    print("3. ค้นหารายชื่อนักศึกษา (Search)")
    print("4. เรียงลำดับรายชื่อ (Sort)")
    print("5. ออกจากโปรแกรม (Exit)")

    # รับค่าตัวเลือกจากผู้ใช้
    choice = input("กรุณาเลือกเมนู (1-5): ")

    # ============================================
    # ตัวเลือกที่ 1: เพิ่มรายชื่อนักศึกษา
    # ============================================
    if choice == "1":
        # รับชื่อนักศึกษาจากผู้ใช้
        name = input("กรอกชื่อนักศึกษาที่ต้องการเพิ่ม: ")
        # ใช้ append() เพิ่มชื่อเข้าไปต่อท้าย list
        student_names.append(name)
        print(f"✅ เพิ่ม '{name}' เรียบร้อยแล้ว")
        # แสดงรายชื่อปัจจุบันทั้งหมด
        print(f"📋 รายชื่อปัจจุบัน: {student_names}")
        print(f"📊 จำนวนทั้งหมด: {len(student_names)} คน")

    # ============================================
    # ตัวเลือกที่ 2: ลบรายชื่อนักศึกษา
    # ============================================
    elif choice == "2":
        # เช็คก่อนว่ามีข้อมูลใน list หรือไม่
        if len(student_names) == 0:
            print("⚠️ ไม่มีรายชื่อนักศึกษาในระบบ กรุณาเพิ่มก่อน")
        else:
            print(f"📋 รายชื่อปัจจุบัน: {student_names}")
            name = input("กรอกชื่อที่ต้องการลบ: ")
            # ต้องเช็คก่อนว่ามีชื่อนี้อยู่ใน list จริงหรือไม่
            # ถ้าไม่เช็คก่อนแล้วสั่ง remove() จะเกิด ValueError
            if name in student_names:
                student_names.remove(name)  # ลบชื่อออกจาก list
                print(f"✅ ลบ '{name}' เรียบร้อยแล้ว")
            else:
                print(f"❌ ไม่พบชื่อ '{name}' ในรายชื่อ")
            print(f"📋 รายชื่อปัจจุบัน: {student_names}")

    # ============================================
    # ตัวเลือกที่ 3: ค้นหารายชื่อนักศึกษา
    # ============================================
    elif choice == "3":
        if len(student_names) == 0:
            print("⚠️ ไม่มีรายชื่อนักศึกษาในระบบ กรุณาเพิ่มก่อน")
        else:
            name = input("กรอกชื่อที่ต้องการค้นหา: ")
            # ใช้ 'in' ตรวจสอบว่ามีชื่อนี้อยู่ใน list หรือไม่
            if name in student_names:
                # ถ้าเจอ ใช้ index() หาตำแหน่ง
                position = student_names.index(name)
                print(f"✅ พบ '{name}' อยู่ที่ตำแหน่ง index {position}")
            else:
                print(f"❌ ไม่พบชื่อ '{name}' ในรายชื่อ")

    # ============================================
    # ตัวเลือกที่ 4: เรียงลำดับรายชื่อ (A-Z)
    # ============================================
    elif choice == "4":
        if len(student_names) == 0:
            print("⚠️ ไม่มีรายชื่อนักศึกษาในระบบ กรุณาเพิ่มก่อน")
        else:
            # ใช้ sort() เรียงลำดับตามตัวอักษร (A-Z / ก-ฮ)
            student_names.sort()
            print("✅ เรียงลำดับเรียบร้อยแล้ว")
            print(f"📋 รายชื่อหลังเรียงลำดับ: {student_names}")

    # ============================================
    # ตัวเลือกที่ 5: ออกจากโปรแกรม
    # ============================================
    elif choice == "5":
        print("\n👋 ขอบคุณที่ใช้งาน ออกจากโปรแกรมเรียบร้อยแล้ว")
        print(f"📋 รายชื่อสุดท้าย: {student_names}")
        print(f"📊 จำนวนนักศึกษาทั้งหมด: {len(student_names)} คน")
        break  # ใช้ break เพื่อหยุด while loop ออกจากโปรแกรม

    # ============================================
    # กรณีกรอกตัวเลือกไม่ถูกต้อง
    # ============================================
    else:
        print("❌ ตัวเลือกไม่ถูกต้อง กรุณาเลือก 1-5 เท่านั้น")
```

# lab4_2.py - Geometric Calculations (Tuples & Immutability)

```python
# lab4_2.py - Geometric Calculations (Tuples & Immutability)
# หมายเหตุ: โค้ดนี้ตัดส่วน "Challenge" (รับพิกัดจากผู้ใช้เพื่อคำนวณระยะทาง) ออก
#          เพื่อให้นักศึกษานำ concept ด้านล่างไปต่อยอดคิดเองเป็นแบบฝึกหัด

import math

print("--- Lab 4.2: Geometric Calculations ---")

# 1) กำหนดจุดสองจุดด้วย tuple: point1 และ point2
point1 = (3, 4) #เติมพิกัด x y ของจุด 👈ต้องทำ
point2 = (6, 8)  #เติมพิกัด x y ของจุด 👈ต้องทำ

# 2) เข้าถึงค่าพิกัด x, y ของแต่ละจุด
print("\n[1] Access Coordinates")
print(f"Point 1: x = {point1[0]}, y = {point1[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ
print(f"Point 2: x = {point2[0]}, y = {point2[1]}") #เติมเลข 0 และ 1 ใน [] 👈ต้องทำ

# 3) ทดลองแก้ไขค่าใน tuple เพื่อแสดงให้เห็นว่า tuple เปลี่ยนแปลงไม่ได้ (immutable)
print("\n[2] Attempt Modification (Immutability Demonstration)")
try:
    point1[0] = 5  # บรรทัดนี้จะทำให้เกิด TypeError เพราะ tuple แก้ไขไม่ได้
except TypeError as e:
    print(f"Error trying to modify tuple: {e}")

# 4) คำนวณระยะห่างระหว่างจุดสองจุดด้วยสูตรระยะทาง
#    distance = sqrt( (x2 - x1)^2 + (y2 - y1)^2 )
print("\n[3] Calculate Distance")
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2) #เติมเลขตำแหน่งในลิส 0 และ 1 ใน [] ให้ถูกต้อง 👈ต้องทำ
print(f"Distance between point1 {point1} and point2 {point2}: {distance}")

# --------------------------------------------------------------------
# ให้นักศึกษาลองต่อยอดเอง (Challenge):
# ลองแก้โค้ดให้รับพิกัดของจุดใหม่ 2 จุดจากผู้ใช้ผ่าน input()
# (อย่าลืมแปลงชนิดข้อมูลจาก input() ให้เป็นตัวเลขก่อนนำไปคำนวณ)
# แล้วคำนวณระยะห่างระหว่างจุดทั้งสองที่ผู้ใช้กรอกเข้ามา
# --------------------------------------------------------------------
```

# Challenge Lab 4.2: Geometric Calculations (User Input Version)
โจทย์ท้าทาย: รับพิกัด (x, y) ของจุด 2 จุดจากผู้ใช้ผ่าน input()
โดยต้องแปลงชนิดข้อมูล (Type Conversion) จาก string เป็น float
แล้วนำมาคำนวณหาระยะห่างระหว่างจุดทั้งสอง

```python
# =============================================================
# Challenge Lab 4.2: Geometric Calculations (User Input Version)
# =============================================================
# แนวคิด: แทนที่จะกำหนดค่าพิกัดตายตัวในโปรแกรม (hardcode)
# เราจะให้ผู้ใช้กรอกพิกัด x, y ของจุดทั้ง 2 จุดเอง
#
# สิ่งที่ต้องระวัง:
#   - input() จะคืนค่าเป็น string (ข้อความ) เสมอ
#   - ต้องใช้ float() หรือ int() แปลงเป็นตัวเลขก่อนคำนวณ
#   - ใช้ float() แทน int() เพื่อรองรับค่าทศนิยมได้ด้วย
# =============================================================

import math

print("=" * 50)
print("  Geometric Distance Calculator (User Input)")
print("=" * 50)

# --- รับพิกัดจุดที่ 1 จากผู้ใช้ ---
print("\n📍 กรอกพิกัดจุดที่ 1:")
# input() คืนค่าเป็น string เสมอ
# ดังนั้นต้องใช้ float() ครอบเพื่อแปลงเป็นตัวเลขทศนิยม
x1 = float(input("  x1 = "))
y1 = float(input("  y1 = "))
# สร้าง tuple เก็บพิกัดจุดที่ 1
point1 = (x1, y1)

# --- รับพิกัดจุดที่ 2 จากผู้ใช้ ---
print("\n📍 กรอกพิกัดจุดที่ 2:")
x2 = float(input("  x2 = "))
y2 = float(input("  y2 = "))
# สร้าง tuple เก็บพิกัดจุดที่ 2
point2 = (x2, y2)

# --- แสดงพิกัดที่ผู้ใช้กรอก ---
print(f"\n📌 Point 1: x = {point1[0]}, y = {point1[1]}")
print(f"📌 Point 2: x = {point2[0]}, y = {point2[1]}")

# --- คำนวณระยะห่างด้วยสูตรระยะทาง (Distance Formula) ---
# สูตร: distance = sqrt( (x2 - x1)^2 + (y2 - y1)^2 )
#
# ขั้นตอนการคำนวณ:
#   1. หาผลต่างของ x:  dx = x2 - x1
#   2. หาผลต่างของ y:  dy = y2 - y1
#   3. ยกกำลังสอง:     dx^2 + dy^2
#   4. หารากที่สอง:    sqrt(dx^2 + dy^2)
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)

# --- แสดงผลลัพธ์ ---
print(f"\n📏 ระยะห่างระหว่าง {point1} และ {point2} = {distance:.4f}")
# :.4f หมายถึงแสดงทศนิยม 4 ตำแหน่ง เพื่อความสวยงามของผลลัพธ์
```

# Activity 4.1: List vs. Tuple: Scenario Matching
สถานการณ์ต่อไปนี้ แต่ละเหตุการณ์ควรใช้ list หรือ tuple เพราะอะไร?

**หลักการเลือก:**
- **List** = ข้อมูลที่ **เปลี่ยนแปลงได้** (Mutable) เช่น เพิ่ม ลบ แก้ไขได้
- **Tuple** = ข้อมูลที่ **เปลี่ยนแปลงไม่ได้** (Immutable) เช่น ค่าคงที่ ข้อมูลตายตัว

```python
# =============================================================
# Activity 4.1: List vs. Tuple: Scenario Matching
# =============================================================
# หลักการสำคัญ:
#   List  (Mutable)   -> ข้อมูลที่ต้องการเพิ่ม/ลบ/แก้ไขภายหลังได้
#   Tuple (Immutable) -> ข้อมูลที่คงที่ ไม่ต้องเปลี่ยนแปลง
# =============================================================

print("=" * 60)
print("  Activity 4.1: List vs. Tuple - Scenario Matching")
print("=" * 60)

# ----- สถานการณ์ที่ 1 -----
# ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์
# คำตอบ: ใช้ List
# เหตุผล: อุณหภูมิอาจต้องเพิ่มข้อมูลใหม่ทุกวัน หรืออาจต้อง
#         แก้ไขค่าที่บันทึกผิด ดังนั้นจึงต้องการโครงสร้างที่เปลี่ยนแปลงได้
print("\n1. ข้อมูลอุณหภูมิประจำวันตลอดหนึ่งสัปดาห์")
print("   ➡️ ใช้ List")
print("   เหตุผล: อุณหภูมิต้องเพิ่มข้อมูลใหม่ทุกวัน หรืออาจแก้ไขค่าที่บันทึกผิดได้")
weekly_temp = [30.5, 31.0, 29.8, 32.1, 28.5, 30.0, 31.5]  # ตัวอย่าง
print(f"   ตัวอย่าง: weekly_temp = {weekly_temp}")

# ----- สถานการณ์ที่ 2 -----
# ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น สีแดงคือ (255, 0, 0))
# คำตอบ: ใช้ Tuple
# เหตุผล: ค่าสี RGB ของเฉดสีใดสีหนึ่งเป็นค่าคงที่ตายตัว
#         สีแดงจะเป็น (255, 0, 0) เสมอ ไม่ควรเปลี่ยนแปลง
print("\n2. ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง")
print("   ➡️ ใช้ Tuple")
print("   เหตุผล: ค่าสี RGB เป็นค่าคงที่ตายตัว เช่น สีแดง = (255,0,0) ไม่ควรเปลี่ยน")
red = (255, 0, 0)  # ตัวอย่าง
print(f"   ตัวอย่าง: red = {red}")

# ----- สถานการณ์ที่ 3 -----
# รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้
# คำตอบ: ใช้ List
# เหตุผล: ผู้ใช้ต้องเพิ่มสินค้า (append) หรือลบสินค้า (remove)
#         ออกจากรถเข็นได้ตลอดเวลา จึงต้องใช้โครงสร้างที่แก้ไขได้
print("\n3. รายการสินค้าในรถเข็นชอปปิง")
print("   ➡️ ใช้ List")
print("   เหตุผล: ผู้ใช้ต้องเพิ่ม/ลบสินค้าจากรถเข็นได้ตลอด จึงต้องเปลี่ยนแปลงได้")
cart = ["Laptop", "Mouse", "Keyboard"]  # ตัวอย่าง
print(f"   ตัวอย่าง: cart = {cart}")

# ----- สถานการณ์ที่ 4 -----
# รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)
# คำตอบ: ใช้ Tuple
# เหตุผล: วันในสัปดาห์เป็นข้อมูลคงที่ มี 7 วันเสมอ
#         ไม่มีเหตุผลที่จะต้องเพิ่ม ลบ หรือแก้ไขชื่อวัน
print("\n4. รายชื่อวันในหนึ่งสัปดาห์")
print("   ➡️ ใช้ Tuple")
print("   เหตุผล: วันในสัปดาห์เป็นค่าคงที่ มี 7 วันเสมอ ไม่ต้องเพิ่มหรือลบ")
days = ("จันทร์", "อังคาร", "พุธ", "พฤหัสบดี", "ศุกร์", "เสาร์", "อาทิตย์")  # ตัวอย่าง
print(f"   ตัวอย่าง: days = {days}")

# ----- สถานการณ์ที่ 5 -----
# คะแนนเก็บของนักเรียนจากการส่งงานชุดต่างๆ
# ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้
# คำตอบ: ใช้ List
# เหตุผล: คะแนนอาจมีการเพิ่มจากงานใหม่ หรือแก้ไขคะแนนเดิม
#         ที่บันทึกผิดได้ ต้องการความยืดหยุ่นในการเปลี่ยนแปลง
print("\n5. คะแนนเก็บของนักเรียนจากการส่งงานชุดต่างๆ")
print("   ➡️ ใช้ List")
print("   เหตุผล: คะแนนอาจเพิ่มจากงานใหม่ หรือแก้ไขคะแนนเดิมที่บันทึกผิดได้")
scores = [85, 90, 78, 92]  # ตัวอย่าง
print(f"   ตัวอย่าง: scores = {scores}")

# ----- สถานการณ์ที่ 6 -----
# พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่
# คำตอบ: ใช้ Tuple
# เหตุผล: พิกัด GPS ของสถานที่ที่ตั้งอยู่กับที่ (เช่น อนุสาวรีย์)
#         เป็นค่าคงที่ ไม่มีการย้ายตำแหน่ง จึงไม่ควรเปลี่ยนแปลง
print("\n6. พิกัด GPS ของสถานที่สำคัญ/จุดแลนด์มาร์กที่ตั้งอยู่กับที่")
print("   ➡️ ใช้ Tuple")
print("   เหตุผล: พิกัด GPS ของสถานที่ที่ตั้งอยู่กับที่เป็นค่าคงที่ ไม่มีการย้ายตำแหน่ง")
wat_arun_gps = (13.7437, 100.4888)  # ตัวอย่าง: พิกัดวัดอรุณ กรุงเทพฯ
print(f"   ตัวอย่าง: wat_arun_gps = {wat_arun_gps} (วัดอรุณ กรุงเทพฯ)")

# ----- สรุป -----
print("\n" + "=" * 60)
print("  สรุป: หลักการเลือกใช้ List กับ Tuple")
print("=" * 60)
print("  List  → ข้อมูลที่ต้องเปลี่ยนแปลง (เพิ่ม/ลบ/แก้ไข) ภายหลัง")
print("  Tuple → ข้อมูลที่คงที่ตายตัว ไม่ต้องการเปลี่ยนแปลง")
print("=" * 60)
```

# Activity 4.2: Interactive To-Do List (using List Operations)
พัฒนาโปรแกรมจัดการ To-Do List แบบข้อความอย่างง่าย
โดยใช้ while loop แสดงเมนูให้ผู้ใช้เลือก:
เพิ่มงาน / ดูรายการ / ลบงานที่เสร็จแล้ว / ออกจากโปรแกรม

```python
# =============================================================
# Activity 4.2: Interactive To-Do List
# =============================================================
# แนวคิด: สร้างโปรแกรม To-Do List ที่ให้ผู้ใช้จัดการรายการงาน
# ที่ต้องทำผ่านเมนูคำสั่ง โดยใช้คำสั่งจัดการ List:
#   - append() : เพิ่มงานใหม่
#   - pop()    : ลบงานตาม index
#   - remove() : ลบงานตามชื่อ
#   - len()    : นับจำนวนงานทั้งหมด
# =============================================================

print("=" * 50)
print("  📝 Interactive To-Do List")
print("=" * 50)

# สร้าง list ว่างสำหรับเก็บรายการงานที่ต้องทำ
todo_list = []

# ใช้ while True วนลูปแสดงเมนูไปเรื่อยๆ
while True:
    # แสดงเมนูคำสั่ง
    print("\n--- 📋 เมนู To-Do List ---")
    print("1. เพิ่มรายการงาน (Add a task)")
    print("2. ดูรายการงานทั้งหมด (View tasks)")
    print("3. ทำเครื่องหมายงานเสร็จ (Mark as done / Remove)")
    print("4. ออกจากโปรแกรม (Exit)")

    choice = input("กรุณาเลือกเมนู (1-4): ")

    # ============================================
    # ตัวเลือกที่ 1: เพิ่มรายการงาน
    # ============================================
    if choice == "1":
        task = input("กรอกรายการงานที่ต้องทำ: ")
        # ใช้ append() เพิ่มงานเข้าไปต่อท้าย list
        todo_list.append(task)
        print(f"✅ เพิ่มงาน '{task}' เรียบร้อยแล้ว")
        print(f"📊 จำนวนงานทั้งหมด: {len(todo_list)} รายการ")

    # ============================================
    # ตัวเลือกที่ 2: ดูรายการงานทั้งหมด
    # ============================================
    elif choice == "2":
        if len(todo_list) == 0:
            print("📭 ไม่มีรายการงาน — ว่างเปล่า!")
        else:
            print(f"\n📋 รายการงานทั้งหมด ({len(todo_list)} รายการ):")
            # ใช้ enumerate() เพื่อแสดง index คู่กับชื่องาน
            # enumerate() จะคืนค่าเป็นคู่ (index, value)
            for index, task in enumerate(todo_list):
                print(f"   [{index}] {task}")

    # ============================================
    # ตัวเลือกที่ 3: ทำเครื่องหมายงานเสร็จ (ลบงาน)
    # ============================================
    elif choice == "3":
        if len(todo_list) == 0:
            print("📭 ไม่มีรายการงาน — ไม่มีอะไรให้ลบ")
        else:
            # แสดงรายการงานทั้งหมดพร้อม index ให้ผู้ใช้เลือก
            print(f"\n📋 รายการงานปัจจุบัน:")
            for index, task in enumerate(todo_list):
                print(f"   [{index}] {task}")

            # ให้ผู้ใช้เลือกวิธีลบ: ตาม index หรือตามชื่อ
            print("\nเลือกวิธีลบ:")
            print("  a = ลบตามตำแหน่ง (index)")
            print("  b = ลบตามชื่องาน")
            method = input("เลือก (a/b): ")

            if method == "a":
                # ลบตาม index โดยใช้ pop()
                try:
                    idx = int(input("กรอกหมายเลข index ที่ต้องการลบ: "))
                    # pop() จะลบสมาชิกที่ตำแหน่ง index และคืนค่าที่ถูกลบกลับมา
                    removed_task = todo_list.pop(idx)
                    print(f"✅ งาน '{removed_task}' เสร็จแล้ว! ลบออกเรียบร้อย")
                except IndexError:
                    # เกิดเมื่อ index ที่กรอกอยู่นอกช่วงของ list
                    print("❌ หมายเลข index ไม่ถูกต้อง กรุณาลองใหม่")
                except ValueError:
                    # เกิดเมื่อผู้ใช้กรอกค่าที่ไม่ใช่ตัวเลข
                    print("❌ กรุณากรอกเป็นตัวเลขเท่านั้น")

            elif method == "b":
                # ลบตามชื่อ โดยใช้ remove()
                task_name = input("กรอกชื่องานที่ต้องการลบ: ")
                if task_name in todo_list:
                    todo_list.remove(task_name)
                    print(f"✅ งาน '{task_name}' เสร็จแล้ว! ลบออกเรียบร้อย")
                else:
                    print(f"❌ ไม่พบงาน '{task_name}' ในรายการ")
            else:
                print("❌ ตัวเลือกไม่ถูกต้อง กรุณาเลือก a หรือ b")

            # แสดงรายการที่เหลือ
            print(f"📊 งานที่เหลือ: {len(todo_list)} รายการ")

    # ============================================
    # ตัวเลือกที่ 4: ออกจากโปรแกรม
    # ============================================
    elif choice == "4":
        print("\n👋 ขอบคุณที่ใช้งาน To-Do List!")
        if len(todo_list) > 0:
            print(f"⚠️ ยังมีงานค้างอยู่ {len(todo_list)} รายการ:")
            for index, task in enumerate(todo_list):
                print(f"   [{index}] {task}")
        else:
            print("🎉 ทำงานครบหมดแล้ว เก่งมาก!")
        break  # หยุด while loop ออกจากโปรแกรม

    # ============================================
    # กรณีกรอกตัวเลือกไม่ถูกต้อง
    # ============================================
    else:
        print("❌ ตัวเลือกไม่ถูกต้อง กรุณาเลือก 1-4 เท่านั้น")
```
